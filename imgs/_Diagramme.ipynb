```python
bar = alt.Chart(cpu_idle_gesamt).mark_bar().encode(
    alt.X("value", scale=alt.Scale(domain=[cpu_idle_gesamt.value.min(), cpu_idle_gesamt.value.max()])),
    alt.Y("epoch",scale=alt.Scale(domain=[cpu_idle_gesamt.epoch.min(), cpu_idle_gesamt.epoch.max()])),
    color = "CPU"
).interactive().properties(width=400)

area = alt.Chart(cpu_idle_gesamt).mark_area(opacity=0.3).encode(
    alt.X("value", scale=alt.Scale(domain=[cpu_idle_gesamt.value.min(), cpu_idle_gesamt.value.max()])),
    alt.Y("epoch",scale=alt.Scale(domain=[cpu_idle_gesamt.epoch.min(), cpu_idle_gesamt.epoch.max()])),
    color="CPU"
).interactive().properties(width=400)

point = alt.Chart(cpu_idle_gesamt).mark_point().encode(
    alt.X("value", scale=alt.Scale(domain=[cpu_idle_gesamt.value.min(), cpu_idle_gesamt.value.max()])),
    alt.Y("epoch",scale=alt.Scale(domain=[cpu_idle_gesamt.epoch.min(), cpu_idle_gesamt.epoch.max()])),
    color = "CPU"
).interactive().properties(width=400)

((bar & point) & area).resolve_scale(color='independent' )
```
