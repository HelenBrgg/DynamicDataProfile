```python
cpu0_idle_l = pd.read_csv("C:/Users/ilove/OneDrive/Uni/Master - Philipps Uni/1. Semester/Projektarbeit/CSV-Dateien Auslastung Felix/light-browsing/cpu-0/percent-idle-2022-08-02")
cpu1_idle_l = pd.read_csv("C:/Users/ilove/OneDrive/Uni/Master - Philipps Uni/1. Semester/Projektarbeit/CSV-Dateien Auslastung Felix/light-browsing/cpu-1/percent-idle-2022-08-02")
cpu2_idle_l = pd.read_csv("C:/Users/ilove/OneDrive/Uni/Master - Philipps Uni/1. Semester/Projektarbeit/CSV-Dateien Auslastung Felix/light-browsing/cpu-2/percent-idle-2022-08-02")
cpu3_idle_l = pd.read_csv("C:/Users/ilove/OneDrive/Uni/Master - Philipps Uni/1. Semester/Projektarbeit/CSV-Dateien Auslastung Felix/light-browsing/cpu-3/percent-idle-2022-08-02")
cpu4_idle_l = pd.read_csv("C:/Users/ilove/OneDrive/Uni/Master - Philipps Uni/1. Semester/Projektarbeit/CSV-Dateien Auslastung Felix/light-browsing/cpu-4/percent-idle-2022-08-02")
cpu5_idle_l = pd.read_csv("C:/Users/ilove/OneDrive/Uni/Master - Philipps Uni/1. Semester/Projektarbeit/CSV-Dateien Auslastung Felix/light-browsing/cpu-5/percent-idle-2022-08-02")
cpu6_idle_l = pd.read_csv("C:/Users/ilove/OneDrive/Uni/Master - Philipps Uni/1. Semester/Projektarbeit/CSV-Dateien Auslastung Felix/light-browsing/cpu-6/percent-idle-2022-08-02")
cpu7_idle_l = pd.read_csv("C:/Users/ilove/OneDrive/Uni/Master - Philipps Uni/1. Semester/Projektarbeit/CSV-Dateien Auslastung Felix/light-browsing/cpu-7/percent-idle-2022-08-02")

cpu0_idle_l = cpu0_idle_l.assign(CPU = 'cpu0_idle')
cpu1_idle_l = cpu1_idle_l.assign(CPU = 'cpu1_idle')
cpu2_idle_l = cpu2_idle_l.assign(CPU = 'cpu2_idle')
cpu3_idle_l = cpu3_idle_l.assign(CPU = 'cpu3_idle')
cpu4_idle_l = cpu4_idle_l.assign(CPU = 'cpu4_idle')
cpu5_idle_l = cpu5_idle_l.assign(CPU = 'cpu5_idle')
cpu6_idle_l = cpu6_idle_l.assign(CPU = 'cpu6_idle')
cpu7_idle_l = cpu7_idle_l.assign(CPU = 'cpu7_idle')

frames_idle_l = [cpu0_idle_l,cpu1_idle_l,cpu2_idle_l,cpu3_idle_l,cpu4_idle_l,cpu5_idle_l,cpu6_idle_l,cpu7_idle_l]
cpu_idle_gesamt_l = pd.concat(frames_idle_l)


memory_used_l = pd.read_csv("C:/Users/ilove/OneDrive/Uni/Master - Philipps Uni/1. Semester/Projektarbeit/CSV-Dateien Auslastung Felix/light-browsing/memory/percent-used-2022-08-02")




mean_workload_l = alt.Chart(cpu_idle_gesamt_l).mark_line().transform_calculate(
    workload='100 - datum.value',
    ).encode(
        alt.Y("workload:Q", scale=alt.Scale(domain=[100-cpu_idle_gesamt_l.value.max(), 100-cpu_idle_gesamt_l.value.min()]), aggregate=alt.Aggregate("mean"), title="Mean Workload"),
        alt.X("epoch", scale=alt.Scale(domain=[cpu_idle_gesamt_l.epoch.min(), cpu_idle_gesamt_l.epoch.max()]), title="Timestamp"),
        color = alt.value("#FFAA00")
).interactive().properties(width=400)

workload_l = alt.Chart(cpu_idle_gesamt_l).mark_line().transform_calculate(
    a='100 - datum.value',
    ).encode(
        alt.Y("a:Q", scale=alt.Scale(domain=[100-cpu_idle_gesamt_l.value.max(), 100-cpu_idle_gesamt_l.value.min()]), title="Workload per CPU"),
        alt.X("epoch", scale=alt.Scale(domain=[cpu_idle_gesamt_l.epoch.min(), cpu_idle_gesamt_l.epoch.max()]), title="Timestamp"),
        color = "CPU"
).interactive().properties(width=400)

memory_l = alt.Chart(memory_used_l).mark_line().encode(
        y = alt.Y("value", scale=alt.Scale(domain=[memory_used_l.value.min(), memory_used_l.value.max()]), title="Memory Use"),
        x = alt.X("epoch", scale=alt.Scale(domain=[memory_used_l.epoch.min(), memory_used_l.epoch.max()]), title="Timestamp"),
).interactive().properties(width=400)



light_browsing = alt.hconcat(mean_workload_l + memory_l, memory_l,
  title='Mean Workload and Memory Usage with light browsing',
).resolve_legend(
  color='independent'
).resolve_scale(color='independent')

light_browsing
```
